Configure gemini

```python
try:
 from google import genai
 from google.colab import userdata
 import numpy as np
 client=genai.Client(api_key=userdata.get("geminikey"))
 print("load succesfully!")
except ImportError:
  print("Error:GOOGLE.COLAB NOT AVAILABLE")
except Exception as e:
  print("Error loading Gemini Key:",e)
```

Produced natural language explanation for given prompt

```python
try:
    sentences = [
        "Python is a programming language",
        "Java is used for enterprise applications",
        "Machine learning is a branch of AI",
        "Neural networks are inspired by the brain",
        "Dogs are loyal animals",
        "Cats like sleeping",
        "Football is a popular sport",
        "Cricket is played in India",
        "Databases store information",
        "SQL queries databases",
        "Operating systems manage hardware",
        "Linux is an operating system",
        "Cloud computing provides remote resources",
        "Cybersecurity protects systems",
        "Data science uses statistics",
        "Algorithms solve problems",
        "Networks connect computers",
        "Routers forward packets",
        "Web development creates websites",
        "HTML structures web pages"
    ]
    print(f"Defined {len(sentences)} sentences")
except Exception as e:
    print("Error defining sentences:")
    print(e)
```

```python
try:
    embeddings = []

    for sentence in sentences:
        response = client.models.embed_content(
            model="gemini-embedding-001",
            contents=sentence
        )
        embeddings.append(response.embeddings[0].values)

    print("Embeddings generated successfully")
    print("Number of embeddings:", len(embeddings))
    print("Embedding dimension:", len(embeddings[0]))

except KeyError as e:
    print("KeyError - Check if 'embeddings' key exists in response:")
    print(e)
except AttributeError as e:
    print("AttributeError - Check if 'models' or 'embed_content' exists:")
    print(e)
except Exception as e:
    print("Unexpected error while generating embeddings:")
    print(e)
```

```python
try:
    embeddings = np.array(embeddings)
    print("Embeddings shape:", embeddings.shape)
except NameError as e:
    print("NameError - 'embeddings' not defined:")
    print(e)
except Exception as e:
    print("Unexpected error converting to numpy array:")
    print(e)
```

Converted text sentence into numerical vector representation using 20 sentences.

```python
try:
    query = "Explain python"

    query_response = client.models.embed_content(
        model="gemini-embedding-001",
        contents=query
    )
    query_embedding = np.array(query_response.embeddings[0].values)

    print("Query embedding generated successfully!")
    print("Query embedding shape:", query_embedding.shape)

except KeyError as e:
    print("KeyError - Check if 'embeddings' key exists in query response:")
    print(e)
except AttributeError as e:
    print("AttributeError - Check client configuration:")
    print(e)
except Exception as e:
    print("Unexpected error generating query embedding:")
    print(e)
```

```python
try:
    from sklearn.metrics.pairwise import cosine_similarity
    similarities = cosine_similarity([query_embedding], embeddings)
    top_5_indices = similarities[0].argsort()[-5:][::-1]

    print("Query:")
    print(query)
    print("\nTop 5 Similar Sentences:\n")

    for idx in top_5_indices:
        print(sentences[idx])
        print("Similarity Score:", round(similarities[0][idx], 4))
        print()

except NameError as e:
    print("NameError - Check if variables are defined:")
    print(" - query_embedding:", 'query_embedding' in locals())
    print(" - embeddings:", 'embeddings' in locals())
    print(" - sentences:", 'sentences' in locals())
    print(e)
except ValueError as e:
    print("ValueError - Check if embeddings shapes match:")
    print("Query embedding shape:", query_embedding.shape)
    print("Sentences embeddings shape:", embeddings.shape)
    print(e)
except IndexError as e:
    print("IndexError - Check if top_5_indices are within range:")
    print(e)
except Exception as e:
    print("Unexpected error during similarity search:")
    print(e)
```

Cosine Similarly search.retrieve top 5 similar sentence to the given query from the earlier embedded 20 sentence.
