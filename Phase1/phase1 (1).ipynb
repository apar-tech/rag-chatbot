```python
from groq import Groq
from google.colab import userdata

try:
    client = Groq(
        api_key=userdata.get("groqkey")
    )

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": "What is RAG?"
            }
        ]
    )

    print(response.choices[0].message.content)

except Exception as e:
    print("Error occurred:", e)
```

```python
import google.generativeai as genai
from google.colab import userdata
import numpy as np
genai.configure(api_key=userdata.get("geminikey"))
```

Produced natural language explanation for given prompt

```python
try:
    embeddings = []

    sentences = [
        "Python is a programming language",
        "Java is used for enterprise applications",
        "Machine learning is a branch of AI",
        "Neural networks are inspired by the brain",
        "Dogs are loyal animals",
        "Cats like sleeping",
        "Football is a popular sport",
        "Cricket is played in India",
        "Databases store information",
        "SQL queries databases",
        "Operating systems manage hardware",
        "Linux is an operating system",
        "Cloud computing provides remote resources",
        "Cybersecurity protects systems",
        "Data science uses statistics",
        "Algorithms solve problems",
        "Networks connect computers",
        "Routers forward packets",
        "Web development creates websites",
        "HTML structures web pages"
    ]

    for sentence in sentences:
        response = client.models.embed_content(
            model="gemini-embedding-001",
            contents=sentence
        )

        embeddings.append(response.embeddings[0].values)

    print("Embeddings generated successfully")
    print("Number of embeddings:", len(embeddings))
    print("Embedding dimension:", len(embeddings[0]))

except Exception as e:
    print("Error while generating embeddings:")
    print(e)
```

```python
embeddings=np.array(embeddings)
print("Embeddings shapes:",embeddings.shape)
```

Converted text sentence into numerical vector representation using 20 sentences.

```python
try:
  query = "Explain python"

  query_response = client.models.embed_content(
      model="gemini-embedding-001",
      contents=query
  )
  query_embedding=np.array(query_response.embeddings[0].values)

  print("Query embedding generated succesfully!")
except Exception as e:
  print("Error while generating query embedding:")
  print(e)
```

```python
try:
    from sklearn.metrics.pairwise import cosine_similarity

    similarities = cosine_similarity(
        [query_embedding],
        embeddings
    )

    # Get top 5 most similar sentences
    top_5_indices = similarities[0].argsort()[-5:][::-1]

    print("Query:")
    print(query)

    print("\nTop 5 Similar Sentences:\n")

    for idx in top_5_indices:
        print(sentences[idx])
        print("Similarity Score:", round(similarities[0][idx], 4))
        print()

except Exception as e:
    print("Error during similarity search:")
    print(e)
```
