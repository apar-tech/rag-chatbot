```python

from langchain_google_genai import GoogleGenerativeAIEmbeddings

try:
    embeddings = GoogleGenerativeAIEmbeddings(
        model="gemini-embedding-001",
    )


    sentences = [
        "Python is a programming language",
        "Java is used for enterprise applications",
        "Machine learning is a branch of AI",
        "Neural networks are inspired by the brain",
        "Dogs are loyal animals",
        "Cats like sleeping",
        "Football is a popular sport",
        "Cricket is played in India",
        "Databases store information",
        "SQL queries databases",
        "Operating systems manage hardware",
        "Linux is an operating system",
        "Cloud computing provides remote resources",
        "Cybersecurity protects systems",
        "Data science uses statistics",
        "Algorithms solve problems",
        "Networks connect computers",
        "Routers forward packets",
        "Web development creates websites",
        "HTML structures web pages"
    ]

    vectors = embeddings.embed_documents(sentences)

    print("Total embeddings:", len(vectors))
    print("Dimension:", len(vectors[0]))

except Exception as e:
    print("Error occurred while generating embeddings:", e)
```

Converted text sentence into numerical vector representation using 20 sentences.

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

try:
    query = "Artificial Intelligence"

    query_embedding = embeddings.embed_query(query)

    similarities = cosine_similarity(
        [query_embedding],
        vectors
    )[0]

    top_5 = np.argsort(similarities)[-5:][::-1]

    print("Top 5 Similar Sentences:\n")

    for idx in top_5:
        print(f"Sentence: {sentences[idx]}")
        print(f"Similarity: {similarities[idx]:.4f}\n")

except Exception as e:
    print("Error occurred during similarity search:", e)
```

Cosine Similarly search.retrieve top 5 similar sentence to the given query from the earlier embedded 20 sentence.
